```python
import pandas as pd

happiness_df = pd.read_excel("Happiness_Report_2022.xls", usecols=["RANK", "Country", "Happiness score"])
happiness_df.columns = [col_name.lower().replace(" ", "_") for col_name in happiness_df.columns]
happiness_df = happiness_df.rename(columns={"happiness_score": "score"})[["country", "rank", "score"]]
happiness_df.head()
```
